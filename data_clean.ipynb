```python
import pandas as pd
salary_url = "https://raw.github.com/ldeleong/DC_Predictions/main/dc%20income%20.csv"
HousingCost_URL = "https://raw.githubusercontent.com/ldeleong/DC_Predictions/main/dc_Rent_cpi.csv"
food_url = "https://raw.githubusercontent.com/ldeleong/DC_Predictions/main/DC_Food.csv"
trans_URL = "https://raw.github.com/ldeleong/DC_Predictions/main/dc_Transportation_cpi.csv"
utilities_URL = "https://raw.github.com/ldeleong/DC_Predictions/main/DC_Utilities_CPI.csv"
edu_com_URL = "https://raw.github.com/ldeleong/DC_Predictions/main/EDU%26COM_CPI.csv"
furnish_URL = "https://raw.github.com/ldeleong/DC_Predictions/main/Furnish_CPI.csv"
REc_URL = "https://raw.github.com/ldeleong/DC_Predictions/main/REC_CPI.csv"

df_income = pd.read_csv(salary_url)
df_rent = pd.read_csv(HousingCost_URL)
df_Food = pd.read_csv(food_url)
df_trans = pd.read_csv(trans_URL)
df_utilities = pd.read_csv(utilities_URL)
df_EduCOM = pd.read_csv(edu_com_URL)
df_furnish = pd.read_csv(furnish_URL)
df_rec = pd.read_csv(REc_URL)
```

```python
df_income
df_income.rename(columns= {
    'MEHOINUSDCA672N': 'income'
}, inplace=True)

df_income
```

```python
df_income['Year'] = pd.DatetimeIndex(df_income['DATE']).year
df_income
```

```python
DC_income = df_income.copy()
DC_income = DC_income[["Year", "income"]]
DC_income
```

```python
df_rent["AVG_Rent_CPI"] = (df_rent["Jan"] + df_rent["Feb"] + df_rent["Mar"] + df_rent["Apr"] + df_rent["May"] + df_rent["Jun"] + df_rent["Jul"] + df_rent["Aug"] + df_rent["Sep"] + df_rent["Oct"] + df_rent["Nov"] + df_rent["Dec"]) / 12.0
df_rent_average = df_rent.groupby("Year").agg(average=("AVG_Rent_CPI", "mean"))
df_rent
```

```python
df_rent["AVG_CPI"] = (df_rent["Jan"] + df_rent["Feb"] + df_rent["Mar"] + df_rent["Apr"] + df_rent["May"] + df_rent["Jun"] + df_rent["Jul"] + df_rent["Aug"] + df_rent["Sep"] + df_rent["Oct"] + df_rent["Nov"] + df_rent["Dec"]) / 12.0
df_rent.groupby("Year").agg(average=("AVG_CPI", "mean"))
df_rent
```

```python
DC_rent = df_rent.copy()
DC_rent = DC_rent[["Year", "AVG_Rent_CPI"]]
DC_rent
```

```python
df_Food = df_Food.loc[192: 215]
df_Food = df_Food[(1997 <= df_Food['Year']) & (df_Food['Year'] <= 2020)]
df_Food
```

```python
DC_Food = df_Food.copy()
DC_Food = DC_Food[["Year","State","Total nominal food sales per capita","Total constant dollar food sales per capita"]]
del DC_Food["State"]
DC_Food
```

```python
df_trans["AVG_trans_CPI"] = (df_trans["Jan"] + df_trans["Mar"] +df_trans["May"] + df_trans["Jul"] + df_trans["Sep"] + df_trans["Nov"] ) / 6.0
df_trans_average = df_trans.groupby("Year").agg(average=("AVG_trans_CPI", "mean"))
df_trans
```

```python
DC_trans = df_trans.copy()
DC_trans = DC_trans[["Year", "AVG_trans_CPI"]]
DC_trans
```

```python
df_utilities["AVG_utilities_CPI"] = (df_utilities["Jan"] + df_utilities["Mar"] +df_utilities["May"] + df_utilities["Jul"] + df_utilities["Sep"] + df_utilities["Nov"] ) / 6.0
df_utilities_average = df_utilities.groupby("Year").agg(average=("AVG_utilities_CPI", "mean"))
df_utilities
```

```python
DC_utilities= df_utilities.copy()
DC_utilities = DC_utilities[["Year", "AVG_utilities_CPI"]]
DC_utilities
```

```python
df_EduCOM["AVG_EduCom_CPI"] = (df_EduCOM["Jan"] + df_EduCOM["Mar"] +df_EduCOM["May"] + df_EduCOM["Jul"] + df_EduCOM["Sep"] + df_EduCOM["Nov"] ) / 6.0
df_EduCom_average = df_EduCOM.groupby("Year").agg(average=("AVG_EduCom_CPI", "mean"))
df_EduCOM
```

```python
DC_EduCom = df_EduCOM.copy()
DC_EduCom = DC_EduCom[["Year", "AVG_EduCom_CPI"]]
DC_EduCom
```

```python
df_furnish["AVG_furnish_CPI"] = (df_furnish["Jan"] + df_furnish["Mar"] +df_furnish["May"] + df_furnish["Jul"] + df_furnish["Sep"] + df_furnish["Nov"] ) / 6.0
df_furnish_average = df_furnish.groupby("Year").agg(average=("AVG_furnish_CPI", "mean"))
df_furnish
```

```python
DC_furnish = df_furnish.copy()
DC_furnish = DC_furnish[["Year", "AVG_furnish_CPI"]]
DC_furnish
```

```python
df_rec["AVG_rec_CPI"] = (df_rec["Jan"] + df_rec["Mar"] +df_rec["May"] + df_rec["Jul"] + df_rec["Sep"] + df_rec["Nov"] ) / 6.0
df_rec_average = df_rec.groupby("Year").agg(average=("AVG_rec_CPI", "mean"))
df_rec
```

```python
DC_rec = df_rec.copy()
DC_rec = DC_rec[["Year", "AVG_rec_CPI"]]
DC_rec
```

```python
DC= DC_income.merge(DC_Food, on = 'Year')
DC = DC.merge(DC_EduCom, on = 'Year')
DC = DC.merge(DC_furnish, on = 'Year')
DC = DC.merge(DC_rec, on = 'Year')
DC = DC.merge(DC_rent, on = 'Year')
DC = DC.merge(DC_trans, on = 'Year')
DC = DC.merge(DC_utilities, on = 'Year')

DC
```

```python
DC.to_csv('DC.csv')
```
