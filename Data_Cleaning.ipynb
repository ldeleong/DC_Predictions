loading up the csv income, housing/basic cost, and the food cost files

```python
import pandas as pd

salary_url = "https://raw.githubusercontent.com/ldeleong/DC_Predictions/master/2015-2019_median-household-income.csv"
HousingCost_URL = "https://raw.githubusercontent.com/ldeleong/DC_Predictions/main/housing%20cost.csv"
food_url = "https://raw.githubusercontent.com/ldeleong/DC_Predictions/main/DC_Food.csv"

df_salary = pd.read_csv(salary_url)

df_Housing = pd.read_csv(HousingCost_URL)

df_Food = pd.read_csv(food_url)

```

```python
df_salary['year'] = pd.to_numeric(df_salary['time'], errors='coerce').astype(int)

df_filtered = df_salary[df_salary['time'].between(2015, 2019)]

median_income_by_year = df_filtered.groupby('time')['median_household_income'].median().reset_index(name='median_household_income')
print("Median household income by year (2015-2019):")
print(median_income_by_year)


```

```python
median_income_by_year.rename(columns={'time': 'Year'}, inplace=True)
median_income_by_year
```

cleaning the housing by multiply each by 12 since its an average for a singular month

```python
# finding the years we want
df_Housing = df_Housing[(2015 <= df_Housing['Year']) & (df_Housing['Year'] <= 2019)]

#converting objects to floats
df_Housing['Apartment (3 bedroom) in city centre'] = df_Housing['Apartment (3 bedroom) in city centre'].astype(str)
df_Housing['Apartment (3 bedroom) in city centre'] = df_Housing['Apartment (3 bedroom) in city centre'].str.replace('-', '').astype(float)

df_Housing['Apartment (3 bedroom) oustide city centre'] = df_Housing['Apartment (3 bedroom) oustide city centre'].astype(str)
df_Housing['Apartment (3 bedroom) oustide city centre'] = df_Housing['Apartment (3 bedroom) oustide city centre'].str.replace('-', '').astype(float)
df_Housing

```

```python
DC_Housing = df_Housing
DC_Housing['Apartment (1 bedroom) in city centre'] = DC_Housing['Apartment (1 bedroom) in city centre'] * 12
DC_Housing['Apartment (1 bedroom) oustide city centre'] = DC_Housing['Apartment (1 bedroom) oustide city centre'] * 12
DC_Housing['Apartment (3 bedroom) in city centre'] = DC_Housing['Apartment (3 bedroom) in city centre'] * 12
DC_Housing['Apartment (3 bedroom) oustide city centre'] = DC_Housing['Apartment (3 bedroom) oustide city centre'] * 12
DC_Housing['Basic Apartment Cost '] = DC_Housing['Basic Apartment Cost '] * 12

DC_Housing

```

Here we are cleaning the dc food data since its shows all the food cost of all states in the United States of America database.

We need to find only dc food prices and minimize it by the years 2015 to 2019

after reducing the rows we are reducing the columnns to only see the data that we actually need.

```python
df_Food = df_Food.loc[192:215]
df_Food = df_Food[(2015 <= df_Food['Year']) & (df_Food['Year'] <= 2019)]
df_Food

```

```python
DC_Food = df_Food.copy()
DC_Food = DC_Food[["Year","State","Total nominal food sales per capita","Total constant dollar food sales per capita"]]
del DC_Food["State"]
DC_Food
```

```python
DC= median_income_by_year.merge(DC_Food, on = 'Year')
DC = DC.merge(DC_Housing, on = 'Year')
DC
```

```python
DC.to_csv('DC.csv', index=False)
```
