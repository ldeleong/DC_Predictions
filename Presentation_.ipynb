# **Step 2: Cleaning the data: **

Cleaning the income dataset, we had to give a column name to the income values. Next, we Created the column name “Year”. In the new column “year” it was converted to an integer using the “yyyy” instead of the “yyyy-dd-mm” format. Lastly, we simplified the dataset to only contain the column information of “Year” & “income”.

The second next was cleaning the 6 datasets taken from the U.S bureau. We needed a yearly average, but the data was only providing monthly averages. For the FOOD dataset to clean it, we had to condense the information, so we see the data points from the state “District of Columbia”. Then we extracted information from columns “Year”, “Total nominal food  sales per capita”, and “Total constant dollar food sales per capita”

# **Step 4: Machine learning**

We decided at using  K nearest neighbor (KNN) and linear regression.

Originally, we tested both models we used 7 features. We found that even though all features show a price change on the cost of living it was not the best for our models. ### **KNN:**
For KNN, we carried out a model selection method and chose the best feature combination out of all the 7 features. We then found the best score for the value of K nearest neighbors which was 2.  Finally, we computed the root mean squared error (RMSE) from the cross-validation scores to assess the model's predictive performance. The best RMSE from Root Mean Squared Error was from Cross-Validation: 5517.878722842684 and occoured using 2 nearest neighbors with euclidean distance as the best metric.

```python
import matplotlib.pyplot as plt
years = list(range(2025, 2036))
incomes = [
    101296.95652173914, 103153.91304347826, 105010.86956521739,
    106867.82608695653, 108724.78260869565, 110581.73913043478,
    112438.69565217392, 114295.65217391304, 116152.60869565218,
    118009.5652173913, 119866.52173913043
]
plt.figure(figsize=(10, 6))
plt.plot(years, incomes, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Projected Income')
plt.title('Projected Income from 2025 to 2035')
plt.grid(True)
plt.tight_layout()
plt.show()
```
